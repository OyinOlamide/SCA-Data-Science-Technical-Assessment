BEST-SELLING BOOK OF THE DECADE:

```
data.plot(x = "Year", y = "Price", kind = "scatter", marker = "*");
plt.title("Best-selling books of the decade") 
plt.xlabel("Year")
plt.ylabel("Prices")
```

PERCENTAGE OF FICTION TO NON-FICTION:

```
from collections import Counter
print(Counter(data["Genre"]))
```

```
book_genre = data.groupby("Genre" ).agg("count")
print(book_genre)
```

VISUALIZING PERCENTAGE OF FICTION TO NON-FICTION:

```
import matplotlib.ticker as ticker
import matplotlib.cm as cm
import matplotlib as mpl
from matplotlib.gridspec import GridSpec
import matplotlib.pyplot as plt
label = book_genre.Name.sort_values().index 
count = book_genre.Name.sort_values()
plt.figure(8, figsize=(50, 40)) 
grid = GridSpec(4, 4)
cmap = plt.get_cmap("tab20c")
colors = [cmap(i) for i in np.linspace(0, 6, 10)]
plt.subplot(grid[1, 1], aspect=1, title="Pie presentation of percentage of fiction to non-fiction")
book_type = plt.pie(count, labels=label, autopct="%1.1f%%", shadow=True, colors=colors)
plt.show()
```

RATINGS OVER TIME:

```
data.groupby("Year").mean().plot(y="User Rating");
plt.title("Ratings over time") 
plt.xlabel("Year")
plt.ylabel("User Rating")
```

PRICE DECLINE OF BEST-SELLING BOOK:

```
data.groupby("Year").mean().plot(y="Price");
plt.title("Price decline of best-selling book") 
plt.xlabel("Year")
plt.ylabel("Prices")
```

REVIEWS OF FICTIONAL BOOKS:

```
data[data.Genre == "Fiction"].groupby("Year").mean().plot(y="Reviews");
plt.title("Reviews of fictional books") 
plt.xlabel("Year")
plt.ylabel("# of Reviews")
```

REVIEW OF NON-FICTION BOOKS:

```
data[data.Genre == "Non Fiction"].groupby("Year").mean().plot(y="Reviews");
plt.title("Review of non-fiction books") 
plt.xlabel("Year")
plt.ylabel("# of Reviews")
```

PRICES OF NON-FICTION BOOKS OVER TIME:

```
data[data.Genre == "Non Fiction" ].groupby("Year").mean().plot(y="Price");
plt.title("Prices of non-fiction books over time")
plt.xlabel("Year")
plt.ylabel("Prices")
```

PRICES OF FICTION BOOKS OVER TIME:

```
data[data.Genre == "Fiction"].groupby("Year").mean().plot(y="Price");
plt.title("Prices of fiction books over time") 
plt.xlabel("Year")
plt.ylabel("Prices")
```

RATINGS OF FICTIONAL BOOKS OVER TIME:

```
data[data.Genre == "Fiction"].groupby("Year").mean().plot(y="User Rating");
plt.title("Ratings of fictional books over time") 
plt.xlabel("Year")
plt.ylabel("User rating")
```

RATINGS OF NONFICTION BOOKS OVER TIME:

```
data[data.Genre == "Non Fiction"].groupby("Year").mean().plot(y="User Rating");
plt.title("Ratings of nonfiction books over time") 
plt.xlabel("Year")
plt.ylabel("User Rating")
```

TOP 10 AUTHORS ACCORDING TO REVIEWS:

```
author = data.groupby("Author")["Reviews"].count().reset_index().sort_values("Reviews",ascending=False).head(10).set_index("Author")
plt.subplots(figsize=(20,20))
ax = author["Reviews"].sort_values().plot.barh(width=0.7,color=sns.color_palette("gist_rainbow",12))
ax.set_xlabel("Total Reviews", fontsize=25)
ax.set_ylabel("Authors", fontsize=25)
ax.set_title("Top 10 authors according to reviews",fontsize=30,color="black")
total = []
for i in ax.patches:
    total.append(i.get_width())
total_sum = sum(total)
for i in ax.patches:
    ax.text(i.get_width()+.4, i.get_y()+.4, str(round(i.get_width())), fontsize=20,color="black")
plt.show()
```
